```python
def prerocessing_indo(sentences):
    # case Folding
    sentences = [sentence.lower() for sentence in sentences]

    # tokenizer
    tokenizer     = RegexpTokenizer(r'(?<!\S)[A-Za-z]+(?!\S)|(?<!\S)[A-Za-z]+(?=:(?!\S))')
    # hasil_token   = tokenizer.tokenize(sentences[1])
    token_sentence = [tokenizer.tokenize(sentence) for sentence in sentences]

    # stopword
    # wordlist= set(stopwords.words('indonesian')+["all", "halaman", "off", "mengunggah file"]+ stopwords.words('english'))
    # sw_sentences = [[word for word in sentence if word not in wordlist]for sentence in token_sentence]
    cleaned_sentences = [' '.join(sentence) for sentence in token_sentence]

    sentence = cleaned_sentences
    return sentence
```

```python
news = pd.read_csv("news.csv", on_bad_lines='skip', header= None)
list_news = news.iloc[:,-1].values.tolist()
doc = prerocessing_indo(list_news)
random.shuffle(doc)
doc = [i for i in doc if i]
len(doc)
```

```python
embedding_model = SentenceTransformer('firqaaa/indo-sentence-bert-base')
pretrained_name = "w11wo/indonesian-roberta-base-posp-tagger"

POS_indo = pipeline(
    "token-classification",
    model=pretrained_name,
    tokenizer=pretrained_name
)
```

```python
vectorizer_model = CountVectorizer(stop_words=stopwords.words('indonesian')+["all", "halaman", "off"] + stopwords.words('english'), ngram_range=(1, 2))
```

```python
def objective(trial):
  # embedding_model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
    embedding_model = SentenceTransformer('firqaaa/indo-sentence-bert-base')

    # Parameter UMAP
    n_neighbors = trial.suggest_int('n_neighbors', 20, 150)
    min_dist = trial.suggest_float('min_dist', 0.01, 0.1)


    # Parameter HDBSCAN
    min_cluster_size = trial.suggest_int('min_cluster_size', 20, 150)
    min_samples = trial.suggest_int('min_samples', 20, 80)





    umap_model = umap.UMAP(n_components=2, n_neighbors=n_neighbors, metric="cosine", verbose = True, min_dist= min_dist)

    hdbscan_model = hdbscan.HDBSCAN(min_samples=min_samples, gen_min_span_tree=True, prediction_data=False, min_cluster_size=min_cluster_size)



    representation_models = {
    "POS": POS_indo,
    "KeyBERTInspired": KeyBERTInspired(),
    "MMR": MaximalMarginalRelevance(diversity=0.3),
    "KeyBERT + MMR": [KeyBERTInspired(), MaximalMarginalRelevance(diversity=0.3)],
    # "OpenAI_Label": OpenAI(model="gpt-3.5-turbo", exponential_backoff=True, chat=True, diversity=0.1),
    # "OpenAI_Summary": [KeyBERTInspired(), summarization_model],
    }

    try:
      # Inisialisasi model BertTopic dengan parameter UMAP dan HDBSCAN yang di-suggest oleh Optuna
      model = BERTopic(embedding_model=embedding_model,
                      umap_model=umap_model,
                      hdbscan_model=hdbscan_model,
                      vectorizer_model=vectorizer_model,
                      representation_model=representation_models,
                      )


      # Pelatihan model dengan dataset
      topics, _ = model.fit_transform(doc)

    except Exception as e:
      print(e)
      return 0


    # Evaluasi dengan metrik tertentu, misalnya jumlah cluster yang dihasilkan
    num_clusters = len(set(topics))
    return num_clusters

```

```python
# Buat objek study untuk optimasi dengan Optuna
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=3)  # Jumlah percobaan

# Cetak hasil optimasi
print("Best Parameters:", study.best_params)
print("Best Result:", study.best_value)
```
