# Sentence Transformers
## Kegunaan

Pencarian Semantik: Anda dapat menggunakan representasi vektor yang dihasilkan oleh Sentence Transformers untuk melakukan pencarian semantik yang lebih cerdas. Anda dapat membandingkan teks berdasarkan arti sebenarnya daripada hanya berdasarkan kesamaan kata-kata.

Pemahaman Teks: Sentence Transformers membantu mengatasi masalah dalam pemahaman teks di tingkat makna. Representasi semantik yang kaya memungkinkan Anda untuk memahami lebih baik konteks dan makna yang terkandung dalam kalimat atau paragraf.

Kelompokan dan Klasifikasi: Dengan representasi vektor semantik, Anda dapat melakukan kelompokan atau klasifikasi teks berdasarkan kesamaan makna. Ini bisa sangat berguna dalam analisis data, kategorisasi, atau penandaan otomatis.

Ekstraksi Informasi: Sentence Transformers dapat membantu dalam ekstraksi informasi dari teks dengan mengidentifikasi bagian-bagian teks yang memiliki kesamaan makna atau relevansi.

Pemrosesan Bahasa Alami (NLP): Pustaka ini berguna dalam tugas-tugas pemrosesan bahasa alami seperti pemodelan bahasa, analisis sentimen, penerjemahan, dan tugas lainnya yang melibatkan pemahaman teks.

Pembelajaran Berbasis Representasi: Anda dapat menggunakan model Sentence Transformers sebagai dasar untuk tugas pembelajaran berbasis representasi, di mana Anda menghasilkan representasi vektor yang lebih baik untuk digunakan dalam tugas-tugas lain.

Rekomendasi Konten: Representasi vektor yang semantik memungkinkan sistem rekomendasi untuk lebih baik memahami preferensi pengguna dan merekomendasikan konten yang lebih sesuai.

Analisis Sentimen: Dengan mengambil keuntungan dari pemahaman semantik, Anda dapat menghasilkan analisis sentimen yang lebih baik dan lebih akurat dari teks.

# embeding Text
## Perbedaan

Sentence Transformers juga termasuk dalam kategori "embedding teks" seperti Word2Vec dan TF-IDF. Namun, ada perbedaan penting dalam bagaimana masing-masing metode menghasilkan representasi teks.

Word2Vec: Word2Vec adalah metode yang menghasilkan representasi vektor untuk kata-kata dalam korpus teks. Ini berfokus pada kata-kata individual dan mencoba untuk menangkap makna kontekstual dari kata-kata berdasarkan hubungan antara kata-kata di sekitarnya dalam teks.

TF-IDF (Term Frequency-Inverse Document Frequency): TF-IDF menghasilkan representasi vektor untuk kata-kata dalam dokumen atau teks. Ini memberi bobot pada kata-kata berdasarkan frekuensi kemunculan mereka dalam dokumen tertentu dibandingkan dengan frekuensi kemunculan kata-kata tersebut dalam seluruh korpus.

Sentence Transformers: Sentence Transformers, di sisi lain, menghasilkan representasi vektor untuk kalimat atau paragraf secara keseluruhan. Ini melibatkan lebih dari sekadar kata-kata individual. Model ini memahami konteks dan makna kalimat secara keseluruhan, dan menghasilkan vektor yang merepresentasikan makna semantik dari teks tersebut.

Penting untuk diingat bahwa Word2Vec dan TF-IDF lebih berfokus pada representasi kata-kata, sementara Sentence Transformers lebih fokus pada representasi kalimat atau paragraf. Jadi, jika Anda memerlukan representasi yang lebih tingkat tinggi dan keseluruhan dari teks, Sentence Transformers bisa menjadi pilihan yang lebih baik. Namun, pilihan tergantung pada tugas dan tujuan Anda, serta tingkat detail yang Anda butuhkan dalam representasi teks.

```python
def prerocessing_indo(sentences):
    # case Folding
    sentences = [sentence.lower() for sentence in sentences]

    # tokenizer
    tokenizer     = RegexpTokenizer(r'(?<!\S)[A-Za-z]+(?!\S)|(?<!\S)[A-Za-z]+(?=:(?!\S))')
    # hasil_token   = tokenizer.tokenize(sentences[1])
    token_sentence = [tokenizer.tokenize(sentence) for sentence in sentences]

    # stopword
    wordlist= set(stopwords.words('indonesian')+["all", "halaman", "off", "mengunggah file"])
    sw_sentences = [[word for word in sentence if word not in wordlist]for sentence in token_sentence]
    cleaned_sentences = [' '.join(sentence) for sentence in sw_sentences]

    sentence = cleaned_sentences
    return sentence
```

**Penjelasan tiap paramter**

Document: Ini adalah teks atau isi dokumen yang sedang dinilai. Ini adalah teks aktual dari dokumen dalam kumpulan dokumen yang Anda berikan.

Topic: Nomor atau label topik yang telah diidentifikasi untuk dokumen tertentu. Ini menunjukkan topik yang paling cocok dengan isi dokumen.

Name: Ini mungkin adalah nama atau label yang lebih deskriptif untuk topik yang telah diidentifikasi. Ini bisa membantu memberikan gambaran umum tentang apa yang topik tersebut mewakili.

Representation: Ini mungkin merupakan representasi numerik atau vektor dari dokumen dalam ruang fitur yang dihasilkan oleh model. Representasi semacam itu mungkin membantu dalam perhitungan jarak atau kemiripan antara dokumen.

Representative_Docs: Ini adalah daftar dokumen lain yang dianggap mewakili topik yang sama dengan dokumen yang sedang dievaluasi. Ini bisa membantu Anda melihat dokumen-dokumen lain yang memiliki karakteristik serupa.

Top_n_words: Ini adalah daftar kata-kata yang paling relevan dengan topik yang telah diidentifikasi dalam dokumen. Jumlah kata-kata yang ditampilkan mungkin dibatasi (misalnya, 5 atau 10 kata teratas) untuk memberikan gambaran singkat.

Probability: Ini adalah probabilitas bahwa dokumen tersebut masuk ke dalam topik tertentu. Nilai ini adalah angka antara 0 dan 1 yang menunjukkan sejauh mana dokumen cocok dengan topik tersebut.

Representative_document: Ini adalah salah satu dokumen yang dianggap sebagai perwakilan utama dari topik yang diidentifikasi. Dokumen ini mungkin adalah salah satu dari "Representative_Docs" yang disebutkan di atas.
