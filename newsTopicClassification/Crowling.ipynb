```python
def get_judul_kompas(url):
  listLink_berita = []
  for halaman in range(1, 50):
    politik_url = f"{url}{halaman}"
    req_politik = req.get(politik_url)
    bs_politik = BeautifulSoup(req_politik.text, 'lxml')
    bslist_berita = bs_politik.find_all('div', attrs={'class': 'latest ga--latest mt2 clearfix -newlayout'})
    for berita in bslist_berita:
      for idx in berita.find_all('a'):
        listLink_berita.append(idx.text.strip())
  return listLink_berita
```

```python
def get_link_kompas(url):
  listLink_berita = []
  for halaman in range(1, 5):
    politik_url = f"{url}{halaman}"
    req_politik = req.get(politik_url)
    bs_politik = BeautifulSoup(req_politik.text, 'lxml')
    bslist_berita = bs_politik.find_all('div', attrs={'class': 'latest ga--latest mt2 clearfix -newlayout'})
    for berita in bslist_berita:
      for idx in berita.findAll('a', href = True):
        listLink_berita.append(idx['href'])
  return listLink_berita
```

```python
def text_berita_kompas(list_link:list):
  berita = []
  for link in list_link:
    # url = "https://www.cnnindonesia.com/nasional/20230711122519-32-971989/jokowi-harap-ruu-kesehatan-bisa-penuhi-kekurangan-dokter-dengan-cepat"
    html = req.get(link)
    bs = BeautifulSoup(html.text, 'lxml')
    isi = bs.find_all('div', attrs= {'class': 'read__content'})
    for idx_isi in range(len(isi)):
      isi_html = isi[idx_isi].find_all('p')
      for idx_tx in range(len(isi_html)):
        berita.append(isi_html[idx_tx].text.strip())
        # berita.append(tx.text.strip())
        # print(i.text.strip())
  return berita
```

```python
def get_link(url):
  listLink_berita = []
  for halaman in range(1, 50):
    politik_url = f"{url}{halaman}"
    req_politik = req.get(politik_url)
    bs_politik = BeautifulSoup(req_politik.text, 'lxml')
    bslist_berita = bs_politik.find_all('div', attrs={'class': 'list media_rows middle'})
    for berita in bslist_berita:
      for idx in berita.findAll('a', href = True):
        listLink_berita.append(idx['href'])

  return listLink_berita
```

```python
def title_berita(list_link:list):
  judul = []
  for link in list_link:
    # url = "https://www.cnnindonesia.com/nasional/20230711122519-32-971989/jokowi-harap-ruu-kesehatan-bisa-penuhi-kekurangan-dokter-dengan-cepat"
    html = req.get(link)
    bs = BeautifulSoup(html.text, 'lxml')
    # print(bs.prettify())
    # isi = bs.find_all('p')
    title = bs.find('div', attrs= {'class': 'content_detail'})
    title = title.find('h1').text.strip()
    judul.append(title)
  return judul
```

```python
def text_berita(list_link:list):
  berita = []
  for link in list_link:
    # url = "https://www.cnnindonesia.com/nasional/20230711122519-32-971989/jokowi-harap-ruu-kesehatan-bisa-penuhi-kekurangan-dokter-dengan-cepat"
    html = req.get(link)
    bs = BeautifulSoup(html.text, 'lxml')
    # print(bs.prettify())
    # isi = bs.find_all('p')
    isi = bs.find_all('div', attrs= {'id': 'detikdetailtext'})
    for idx_isi in range(len(isi)):
      isi_html = isi[idx_isi].find_all('p')
      for idx_tx in range(len(isi_html)):
        berita.append(isi_html[idx_tx].text.strip())
        # berita.append(tx.text.strip())
        # print(i.text.strip())
  return berita
```
