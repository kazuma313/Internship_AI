```python
def preprocess(sentence:list):
  # case Folding
  sentences = [sentence.lower() for sentence in sentence]
  # print(sentences)


  # tokenizer
  tokenizer     = RegexpTokenizer(r'\w+')
  # hasil_token   = tokenizer.tokenize(sentences[1])
  token_sentence = [tokenizer.tokenize(sentence) for sentence in sentences]
  # print(token_sentence)

  wordlist= set(stopwords.words('indonesian'))
  sw_sentences = [[word for word in sentence if word not in wordlist]for sentence in token_sentence]
  cleaned_sentences = [' '.join(sentence) for sentence in sw_sentences]
  # print(cleaned_sentences)


  #[str(sentence) for sentence in sw_sentences]
  # stemming
  stemmer     = StemmerFactory().create_stemmer()
  # input_stemm   = str(sw_sentences)
  # hasil_stemm   = stemmer.stem(input_stemm)
  # print (hasil_stemm)
  stem_sentence = [stemmer.stem(sentence) for sentence in cleaned_sentences]
  # print(stem_sentence)
  return stem_sentence
```

```python
def preprocess(sentence:list):
  # case Folding
  sentences = [sentence.lower() for sentence in sentence]
  # print(sentences)


  # tokenizer
  tokenizer     = RegexpTokenizer(r'\w+')
  # hasil_token   = tokenizer.tokenize(sentences[1])
  token_sentence = [tokenizer.tokenize(sentence) for sentence in sentences]
  # print(token_sentence)

  wordlist= set(stopwords.words('indonesian'))
  sw_sentences = [[word for word in sentence if word not in wordlist]for sentence in token_sentence]
  cleaned_sentences = [' '.join(sentence) for sentence in sw_sentences]
  # print(cleaned_sentences)


  #[str(sentence) for sentence in sw_sentences]
  # stemming
  stemmer     = StemmerFactory().create_stemmer()
  # input_stemm   = str(sw_sentences)
  # hasil_stemm   = stemmer.stem(input_stemm)
  # print (hasil_stemm)
  stem_sentence = [stemmer.stem(sentence) for sentence in cleaned_sentences]
  # print(stem_sentence)
  return stem_sentence
```

```python
def berita(kalimat:str):
  kalimat = preprocess([kalimat])
  kalimat = vectorizer.transform(kalimat).toarray()
  pred = modelNN.predict(kalimat)
  length = len([i for i in (pred[0] * 100) if i > 10])

  index_label = np.argsort(pred[0])[::-1][:length]
  pred_label = encoding.inverse_transform(index_label)
  prob_label = pred[0][index_label] * 100

  json = {
    'news' : pred_label.tolist(),
    'probability' : prob_label.tolist()
  }

  return json
```
