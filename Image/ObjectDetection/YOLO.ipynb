```python
def objectDetect(input):
  results = model(input)
  result = results[0]
  # box = result.boxes[0]

  detect = []

  for box in result.boxes:
    class_id = result.names[box.cls[0].item()]
    detect.append(class_id)
    cords = box.xyxy[0].tolist()
    cords = [round(x) for x in cords]
    conf = round(box.conf[0].item(), 2)
    print("Object type:", class_id)
    print("Coordinates:", cords)
    print("Probability:", conf)
    print("---")

  json_detect = {'item' : detect}

  return result, json_detect
```

```python
from fastapi import FastAPI,  File, UploadFile
from fastapi.middleware.cors import CORSMiddleware

import nest_asyncio
from pyngrok import ngrok
import uvicorn
```

```python
!pip install python-multipart
```

```python
import python
```

```python
app = FastAPI()

app.add_middleware(
    CORSMiddleware,
    allow_origins=['*'],
    allow_credentials=True,
    allow_methods=['*'],
    allow_headers=['*'],
)


@app.get('/')
async def root():

  return "This is Object Detection"

@app.post('/ObjectDetecion/url')
async def detect_url(url):
  _, json_data = objectDetect(url)
  return json_data


@app.post('/ObjectDetecion/file')
async def detect_file(file: UploadFile):
  _, json_data = objectDetect(file)
  return json_data
```

```python
ngrok_tunnel = ngrok.connect(8000)
print('Public URL:', ngrok_tunnel.public_url)
nest_asyncio.apply()
uvicorn.run(app, port=8000)
```
